```python
wiki_titles = [
	"Atlanta",
	"Berlin",
	"Boston",
	"Chicago",
	"Houston",
    "Kair", # Cairo
	"Karaczi",
    "Kopenhaga",
	"Lizbona", # Lisbon
	"Londyn", #"London",
	"Moskwa", # "Moscow",
	"Monachium", #"Munich",
	"Pary≈º", # "Paris",
    "Pekin", # Beijing
	"San Francisco",
	"Seattle",
	"Szanghaj", #"Shanghai",
	"Tokio", #"Tokyo",
	"Toronto",
]
```

```python
from pathlib import Path

import requests

for title in wiki_titles:
    response = requests.get(
        "https://pl.wikipedia.org/w/api.php",
        params={
            "action": "query",
            "format": "json",
            "titles": title,
            "prop": "extracts",
            # 'exintro': True,
            "explaintext": True,
        },
    ).json()
    page = next(iter(response["query"]["pages"].values()))
    wiki_text = page["extract"]

    data_path = Path("./polish_city_data")
    if not data_path.exists():
        Path.mkdir(data_path)

    with open(data_path / f"{title}.txt", "w") as fp:
        fp.write(wiki_text)
```
